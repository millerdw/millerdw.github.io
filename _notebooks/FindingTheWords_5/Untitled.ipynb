```python
import pandas as pd
import regex as re
```

```python
with open('meeting_saved_chat.txt','r') as f:
    chat = [(line[:8],line[16:line.find(':', 8)],line[line.find(':', 8)+1:]) for line in f.readlines()]
```

```python
chat = pd.DataFrame(chat, columns=['time','user','message'])
chat
```
